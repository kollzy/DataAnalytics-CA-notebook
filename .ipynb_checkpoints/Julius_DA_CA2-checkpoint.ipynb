```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

# Load the CO2 dataset
co2emissions_df = pd.read_csv('csv/c02_emissions.csv')  

# Clean data 
co2_df = co2emissions_df.dropna(subset=['Year', 'Total CO2 emissions from fossil-fuels and cement production (thousand metric tons of C)'])

# Load the methane emissions dataset
methane_df = pd.read_csv('csv/methane_emissions.csv')  

# Clean data (remove missing values)
methane_df = methane_df.dropna(subset=['Year', 'Annual methane emissions in CO₂ equivalents'])

# Load the deforestation dataset
deforestation_df = pd.read_csv('csv/deforestation.csv') 

# Clean data (remove missing values)
deforestation_df = deforestation_df.dropna(subset=['Year', 'Deforestation'])

# Data Integration
# Group CO2 emissions by year
global_co2 = co2_df.groupby('Year')['Total CO2 emissions from fossil-fuels and cement production (thousand metric tons of C)'].sum().reset_index()

# Group 
global_methane = methane_df.groupby('Year')['Annual methane emissions in CO₂ equivalents'].sum().reset_index()


global_deforestation = deforestation_df.groupby('Year')['Deforestation'].sum().reset_index()

# Merge CO2, methane, and deforestation 
merged_df = pd.merge(global_co2, global_methane, on='Year', how='inner')
merged_df = pd.merge(merged_df, global_deforestation, on='Year', how='inner')

# Select columns
selected_data = merged_df[['Year', 
                           'Total CO2 emissions from fossil-fuels and cement production (thousand metric tons of C)', 
                           'Annual methane emissions in CO₂ equivalents', 
                           'Deforestation']]

# Sort by year
selected_data = selected_data.sort_values(by='Year').reset_index(drop=True)

#Scatter plot 
plt.figure(figsize=(10, 6))
sns.scatterplot(data=selected_data, 
                x='Total CO2 emissions from fossil-fuels and cement production (thousand metric tons of C)', 
                y='Annual methane emissions in CO₂ equivalents', label="CO2 vs Methane")
sns.scatterplot(data=selected_data, 
                x='Deforestation', 
                y='Annual methane emissions in CO₂ equivalents', label="Deforestation vs Methane", color='green')
plt.title('CO2 Emissions, Methane Emissions, and Deforestation')
plt.xlabel('Emissions/Deforestation')
plt.ylabel('Methane Emissions (CO₂ eq.)')
plt.legend()
plt.show()

# Linear Regression model
X = selected_data[['Total CO2 emissions from fossil-fuels and cement production (thousand metric tons of C)', 
                   'Deforestation']]  
y = selected_data['Annual methane emissions in CO₂ equivalents']  


model = LinearRegression()


model.fit(X, y)
# Extract coefficients
intercept = model.intercept_
coefficients = model.coef_


print("\nLinear Regression Model:")
print(f"Intercept (b0): {intercept}")
print(f"Coefficients (b1 for CO2, b2 for Deforestation): {coefficients}")


y_pred = model.predict(X)

#regression fit
plt.figure(figsize=(10, 6))
plt.scatter(selected_data['Total CO2 emissions from fossil-fuels and cement production (thousand metric tons of C)'], 
            selected_data['Annual methane emissions in CO₂ equivalents'], color='blue', alpha=0.7, label='Actual Data')
plt.plot(selected_data['Total CO2 emissions from fossil-fuels and cement production (thousand metric tons of C)'], 
         y_pred, color='red', label='Regression Line')
plt.title('Regression Fit: CO2 Emissions vs Methane Emissions')
plt.xlabel('CO2 Emissions ( CO2)')
plt.ylabel('Methane Emissions (CO2.)')
plt.legend()
plt.show()


plt.figure(figsize=(12, 6))

# CO2 Emissions line plot
plt.subplot(1, 2, 1)
plt.plot(selected_data['Year'], selected_data['Total CO2 emissions from fossil-fuels and cement production (thousand metric tons of C)'], color='blue', label='CO2 Emissions')
plt.title('Global CO2 Emissions Over Time')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions ( CO2)')
plt.legend()

# Deforestation line plot
plt.subplot(1, 2, 2)
plt.plot(selected_data['Year'], selected_data['Deforestation'], color='orange', label='Deforestation')
plt.title('Global Deforestation Over Time')
plt.xlabel('Year')
plt.ylabel('Deforestation')
plt.legend()

plt.tight_layout()
plt.show()
##tried the decision trees but got errors and ran out of time to get it working
```
